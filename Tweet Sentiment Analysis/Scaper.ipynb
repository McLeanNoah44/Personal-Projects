
#This is just the cell for loading in the 3 DFs once you have them saved as CSVs
save_smash_df = pd.read_csv('SaveSmash.csv')
free_melee_df = pd.read_csv('FreeMelee.csv')
mentions_df = pd.read_csv('Mentions.csv')

```python
all_hashtags_df = pd.concat([save_smash_df, free_melee_df])
all_hashtags_df = all_hashtags_df.drop_duplicates(subset=['content'])

print('# of observations in dataset', all_hashtags_df.shape[0])
all_hashtags_df
```

```python
all_hashtags_df['Hashtag Code'] = 'X'

for curr_index, curr_row in all_hashtags_df.iterrows():
    curr_tweet = str.lower(curr_row[2])

    if '#savesmash' in curr_tweet and '#freemelee' in curr_tweet:
        all_hashtags_df.at[curr_index, 'Hashtag Code'] = 'Both'

    elif '#savesmash' in curr_tweet and '#freemelee' not in curr_tweet:
        all_hashtags_df.at[curr_index, 'Hashtag Code'] = 'SaveSmash'

    else:
        all_hashtags_df.at[curr_index, 'Hashtag Code'] = 'FreeMelee'
```

```python
sns.countplot(x= 'Hashtag Code', data = all_hashtags_df)

plt.show()
```

```python
all_hashtags_df['SaveSmash and FreeMelee'] = ['#savesmash' in curr.lower() and '#freesmash' in curr.lower() for curr in all_hashtags_df['content']]

all_hashtags_df['SaveSmash'] = ['#savesmash' in curr.lower() and '#freemelee' not in curr.lower() for curr in all_hashtags_df['content']]

all_hashtags_df['FreeMelee'] = ['#freemelee' in curr.lower() and '#savesmash' not in curr.lower() for curr in all_hashtags_df['content']]

all_hashtags_df['Other Hashtags'] = ['#savesmash' not in curr.lower() and '#freemelee' not in curr.lower() for curr in all_hashtags_df['content']]

mentions_df
```

```python
#Some processing we might not need in the final part
only_save_smash = all_hashtags_df['SaveSmash'].value_counts()[1]
only_free_melee = all_hashtags_df['FreeMelee'].value_counts()[1]
both_hashtags = all_hashtags_df['SaveSmash and FreeMelee'].value_counts()[1]
other_hashtags = all_hashtags_df['Other Hashtags'].value_counts()

print(other_hashtags)

#Plotting
#ax = sns.barplot(x= range(3), y= [only_save_smash, only_free_melee, both_hashtags])
```

```python
import datetime

# Creating a column for the dates (year, month, and day) for each tweet
all_hashtags_df["day"] = all_hashtags_df["date"].apply(lambda x: datetime.datetime.strptime(str(x)[0:10], "%Y-%m-%d"))
# Grouping by each date and summing the hashtags for each day
groups = all_hashtags_df.groupby('day')
both_count = groups['SaveSmash and FreeMelee'].sum()
save_smash_count = groups['SaveSmash'].sum()
free_melee_count = groups['FreeMelee'].sum()

# Getting the list of dates from the group frames
dates = both_count.index.tolist()
# Plotting the number of tweets containing each hashtag over time
sns.lineplot(y=save_smash_count.tolist(), x=dates)
sns.lineplot(y=free_melee_count.tolist(), x=dates)
sns.lineplot(y=both_count.tolist(), x=dates)
plt.xticks(rotation = 45)
plt.legend(labels=["Save Smash","Free Melee", "Both"])
plt.show()
```

```python
# Creating a column for the dates (year, month, and day) for each tweet
mentions_df["day"] = mentions_df["date"].apply(lambda x: datetime.datetime.strptime(str(x)[0:10], "%Y-%m-%d"))
# Grouping by each date and summing the mentions for each day
groups = mentions_df.groupby('day')
mention_count = groups['content'].count()

# Getting the list of dates from the group frame
dates = mention_count.index.tolist()
# Plotting the number of tweets containing a mention over time
sns.lineplot(y=mention_count.tolist(), x=dates)
plt.legend(labels=["Mentions"])
plt.xticks(rotation = 45)
plt.show()
```
